```python
import numpy as np

np.random.seed(0)

m_pos = 50
m_neg = 50
sigma = 0.8

X_pos = sigma * np.random.randn(m_pos, 2) + np.array([2.0, 2.0])
X_neg = sigma * np.random.randn(m_neg, 2) + np.array([-2.0, -2.0])
```

```python
import matplotlib.pyplot as plt 

plt.scatter(X_pos[:,0], X_pos[:,1], color='blue', label='+1 class')
plt.scatter(X_neg[:,0], X_neg[:,1], color='red', label='-1 class')

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Binary Classification Dataset")

plt.legend()
plt.grid(True)

plt.show()
```
